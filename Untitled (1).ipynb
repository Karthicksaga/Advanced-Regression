```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#here there are many columns so we have to saw all the columns
pd.pandas.set_option('display.max_columns',None)
```

```python
train_data=pd.read_csv('train.csv')
train_data
```

```python
train_data.set_index('Id',inplace=True)
train_data
```

```python
train_data.isnull().sum()
```

```python
#finding the features that contain missing values
missing_feature=[features for features in train_data.columns if train_data[features].isnull().sum()>1]
for feature in missing_feature:
    print(feature,np.round(train_data[feature].isnull().mean(),4))
    
```

```python
#there is a strong relationship between missing values and salesprice so we dont just drop the columns
for feature in missing_feature:
    data=train_data.copy()
    
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].mean().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('Salesprice')
    plt.title('comparison of features with salesprice')
    plt.show()
```

```python
#finding the numerical features
numerical_features=[features for features in train_data.columns if train_data[features].dtypes!='O']
print('no. of Numerical features:',len(numerical_features))
train_data[numerical_features].head()
```

```python
#finding the temporal variable in those numerical variable beccause we try to analyse the temporal variable alone
temporal_features=[features for features in numerical_features if 'Yr' in features or 'Year' in features]
temporal_features
```

```python
#so convert it into normal distributiioon using logarithmic transformation
for features in continous_feature:
    data=train_data.copy()
    data[features]=np.log(data[features])
    data["SalePrice"]=np.log(data["SalePrice"])
    plt.scatter(x=data[features],y=data['SalePrice'])
    plt.xlabel(features)
    plt.show()
```

```python
#finding the outliers using the box plot
for features in continous_feature:
    data=train_data.copy()
    data[features]=np.log(data[features])
    data.boxplot(column=features)
    plt.xlabel(features)
    plt.show()
```

```python
#finding and analysis the categorical features
categorical_features=[features for features in train_data.columns if train_data[features].dtypes=='O']
print('no. of categorical features:',len(categorical_features))
train_data[categorical_features].head()
```

```python
for features in categorical_features:
    data=train_data.copy()
    data.groupby(features)['SalePrice'].mean().plot.bar()
    plt.xlabel(features)
    plt.show()
```

```python
#FEATURE ENGINEERING
#first replace the missing values with new label
data=train_data.copy()
data[missing_feature]=data[missing_feature].fillna('missing')
data[missing_feature].isnull().sum()
```

```python
train_data[missing_feature]
```

```python
#handling missing values for numercial variables.here we using median because numerical features have lot of outliers
for features in numerical_features:
    
    median=train_data[features].median()
    
    train_data[features]=train_data[features].fillna(median)
```
